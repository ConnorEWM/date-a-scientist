```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

ages = pd.DataFrame({'Age':df['age']})
y_tr, y_te, x_tr, x_te = train_test_split(words, ages, train_size = 0.8, test_size = 0.2, random_state = 8)
mlr = LinearRegression()
#print(len(x_tr), len(x_te), len(y_tr), len(y_te))
mlr.fit(x_tr, y_tr)
print(mlr.coef_)
predictions = mlr.predict(x_te)

```

```python
x_tr, x_te, y_tr, y_te = train_test_split(ages, words['Total'], train_size = 0.8, random_state = 6)
mlr.fit(x_tr, y_tr)
print(mlr.score(x_te,y_te))

plt.scatter(ages, words['Total'])
plt.plot(x_te, mlr.predict(x_te), 'y')
plt.xlim(18,70)
plt.xlabel('Age')
plt.ylabel('Words in Essay')
plt.show()

plt.plot(x_te, mlr.predict(x_te), 'y')
plt.xlim(18,70)
plt.xlabel('Age')
plt.ylabel('Words in Essay')
plt.show()

predict = mlr.predict(x_te)

plt.plot(range(1,5000),range(1,5000))
plt.scatter(x = y_te.values.tolist(), y = predict, c = 'y')
plt.xlabel('Actual Words')
plt.ylabel('Predicted Words')
plt.show()
```

```python
average = np.sum(words['essay0'])/len(words['essay0'])
stdev = np.std(words['essay0'])

nooutw = []
noouta = []

for i in range(len(words['essay0'])):
    if words.essay0[i] < average + (2 * stdev) and words.essay0[i] > average - (2 * stdev):
        nooutw.append(words.essay0[i])
        noouta.append(ages.Age[i])
#print(nooutw)
#print(noouta)
noa = pd.DataFrame({'Age' : noouta})
now = pd.DataFrame({'Words' : nooutw})

x_tr, x_te, y_tr, y_te = train_test_split(noa, now, train_size = 0.8, random_state = 6)
mlr.fit(x_tr, y_tr)
print(mlr.score(x_te,y_te))

plt.scatter(noouta, nooutw, alpha = 0.1)
plt.plot(x_te, mlr.predict(x_te), 'y')
plt.xlim(18,70)
plt.xlabel('Age')
plt.ylabel('Words in Essay')
plt.show()

predict = mlr.predict(x_te)
print(mlr.predict(x_te))

plt.plot(range(1,2100),range(1,2100), 'y')
plt.xlim(0,5000)
plt.scatter(x = y_te.values.tolist(), y = predict, c = 'b', alpha = 0.1)
plt.xlim(0,5000)
plt.xlabel('Actual Words')
plt.ylabel('Predicted Words')
plt.show()
```
